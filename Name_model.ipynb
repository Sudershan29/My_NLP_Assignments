```python
def gender_feautures(word):
    return{'last_letter':word[-1]}
```

```python
gender_feautures('Obama')
```

```python
from nltk.corpus import names
```

```python
names.words()
```

```python
print(len(names.words())) #Total number of words
```

```python
labeled_names =([{name,'male'}for name in names.words('male.txt')]+[(name,'female') for name in names.words('female.txt')])
```

```python
import random
random.shuffle(labeled_names)
```

```python
feauturessets=[(gender_feautures(n), gender)for (n,gender) in labeled_names]
```

```python
train_set, test_set =feauturessets[6355:], feauturessets[:1589] # spliting the datset in 80% and 20%
```

```python
import nltk
classifier = nltk.NaiveBayesClassifier.train(train_set)
classifier.show_most_informative_features(26)
```

```python
classifier.classify(gender_feautures('Anonymus'))
```

```python
print(nltk.classify.accuracy(classifier,test_set))
```
